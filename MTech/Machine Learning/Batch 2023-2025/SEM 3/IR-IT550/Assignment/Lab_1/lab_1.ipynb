```python
#preprocessing
import pandas as pd
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import re,string
```

```python
df = pd.read_json('lab1-data.json')
```

```python
df
```

```python
new_df = df.drop(columns=['title','summary'])
```

```python
new_df
```

```python
#lowercasing
new_df['text'] = new_df['text'].str.lower()
```

```python
#remove non-ascii chars
new_df['text'] = new_df['text'].str.replace(r'[^\x00-\x7F]+', '',regex=True)
```

```python
#remove html entity code
new_df['text'] = new_df['text'].str.replace(r'&#(\d+);', '',regex=True)
```

```python
#tokenizing
new_df['text'] = new_df['text'].fillna('')
new_df['tokens'] = new_df['text'].apply(word_tokenize)
```

```python
#removing stopwords
stop_words = set(stopwords.words('english'))
new_df['tokens'] = new_df['tokens'].apply(lambda tokens: [word for word in tokens if word.lower() not in stop_words])
```

```python
#removing numbers
new_df['tokens'] = new_df['tokens'].apply(lambda tokens: [word for word in tokens if not word.isdigit() and word not in string.punctuation])
```
