```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mu1 = np.array([-1, -1])
mu2 = np.array([2, 1])
sigma = np.array([[0.7, 0], [0, 0.3]])

# Set the number of data points
n = 2000

# Generate dataset A
np.random.seed(666)  # For reproducibility
data_A = np.random.multivariate_normal(mu1, sigma, n)

# Generate dataset B
np.random.seed(333)  # For reproducibility
data_B = np.random.multivariate_normal(mu2, sigma, n)
```

```python
from sklearn.model_selection import train_test_split
A_train, A_test, B_train, B_test = train_test_split(data_A,data_B, test_size=250, random_state=0)
```

```python
plt.scatter(A_train[:, 0], A_train[:, 1], c='blue', marker='.', label='Dataset A',alpha=0.4)
plt.scatter(B_train[:, 0], B_train[:, 1], c='red', marker='.', label='Dataset B',alpha=0.4)
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Generated Datasets A and B')
plt.show()
```

```python
plt.scatter(A_test[:, 0], A_test[:, 1], c='blue', marker='.', label='Dataset A',alpha=0.4)
plt.scatter(B_test[:, 0], B_test[:, 1], c='red', marker='.', label='Dataset B',alpha=0.4)
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Generated Datasets A and B')
plt.show()
```

```python
sigmaI = np.linalg.inv(sigma)
w=np.dot(sigmaI,(mu1-mu2)).flatten()
b= 0.5 *(np.dot(np.dot(mu1.T, sigmaI),mu1)-np.dot(np.dot(mu2.T,sigmaI),mu2)).flatten()
```

```python
x_values = np.linspace(-2, 4, 400)
y_values = (-w[0] * x_values + b) / w[1]
```

```python
plt.scatter(A_train[:, 0], A_train[:, 1], c='orange', marker='.', label='Dataset A',alpha=0.4)
plt.scatter(B_train[:, 0], B_train[:, 1], c='grey', marker='.', label='Dataset B',alpha=0.4)
plt.plot(x_values, y_values, color='darkblue', linestyle='--', linewidth=2, label='Decision Boundary')
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Generated Datasets A and B')
plt.show()
```

```python
plt.scatter(A_test[:, 0], A_test[:, 1], c='orange', marker='.', label='Dataset A',alpha=0.4)
plt.scatter(B_test[:, 0], B_test[:, 1], c='grey', marker='.', label='Dataset B',alpha=0.4)
plt.plot(x_values, y_values, color='darkblue', linestyle='--', linewidth=2, label='Decision Boundary')
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Generated Datasets A and B')
plt.show()
```
