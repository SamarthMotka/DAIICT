```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mu1 = np.array([0, 0])
mu2 = np.array([2, 2])
sigma1 = np.array([[0.6, 0], [0, 0.6]])  # Identity matrix
sigma2 = np.array([[0.6, 0], [0, 0.6]])  # Identity matrix

# Set the number of data points
n = 2000

# Generate dataset A
np.random.seed(666)  # For reproducibility
data_A = np.random.multivariate_normal(mu1, sigma1, n)

# Generate dataset B
np.random.seed(333)  # For reproducibility
data_B = np.random.multivariate_normal(mu2, sigma2, n)
```

```python
# Optionally, you can visualize the datasets using Matplotlib

plt.scatter(data_A[:, 0], data_A[:, 1], c='blue', marker='.', label='Dataset A',alpha=0.4)
plt.scatter(data_B[:, 0], data_B[:, 1], c='red', marker='.', label='Dataset B',alpha=0.4)
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Generated Datasets A and B')
plt.show()
```

```python
mu1 = mu1.reshape(2,1)
mu2 = mu2.reshape(2,1)

data_A.shape
```

Find the optimal decision boundary for the classification of the dataset A and B using Σ1 = Σ2=

\begin{bmatrix} 0.6 & 0.0 \\ 0.0 & 0.6 \end{bmatrix}


. Plot the dataset A and B with different colors and plot the obtained optimal
decision boundary. Comment on the characteristics of obtained decision boundary.


---



```python
w=(mu1-mu2).T.flatten()
b=np.dot(w,(0.5*(mu1 + mu2))).flatten()

plt.scatter(data_A[:, 0], data_A[:, 1], c='yellow', marker='.', label='Dataset A',alpha=0.4)
plt.scatter(data_B[:, 0], data_B[:, 1], c='skyblue', marker='.', label='Dataset B',alpha=0.4)
x_values = np.linspace(-2, 4, 400)
y_values = (-w[0] * x_values + b) / w[1]
plt.plot(x_values, y_values, color='purple', linestyle='--', linewidth=2, label='Decision Boundary')
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Generated Datasets A and B')
plt.show()
```

Find the optimal decision boundary for the classification of the dataset A and B using Σ1 = Σ2=
\begin{bmatrix} 0.7 & 0.0 \\ 0.0 & 0.3 \end{bmatrix}

Plot the dataset A and B with different colors and plot the obtained optimal
decision boundary. Comment on the characteristics of obtained decision boundary.


---





```python
n = 2000
mu1 = np.array([0, 0])
mu2 = np.array([2, 2])
sigma = np.array([[0.7, 0.0],[0.0, 0.3]])

# Generate dataset A
np.random.seed(66)  # For reproducibility
data_A = np.random.multivariate_normal(mu1, sigma, n)

# Generate dataset B
np.random.seed(33)  # For reproducibility
data_B = np.random.multivariate_normal(mu2, sigma, n)
```

```python

sigmaI = np.linalg.inv(sigma)
w=np.dot(sigmaI,(mu1-mu2)).flatten()
b= 0.5 *(np.dot(np.dot(mu1.T, sigmaI),mu1)-np.dot(np.dot(mu2.T,sigmaI),mu2)).flatten()
```

```python
x_values = np.linspace(-2, 4, 400)
y_values = (-w[0] * x_values + b) / w[1]
```

```python
plt.scatter(data_A[:, 0], data_A[:, 1], c='orange', marker='.', label='Dataset A',alpha=0.4)
plt.scatter(data_B[:, 0], data_B[:, 1], c='grey', marker='.', label='Dataset B',alpha=0.4)
plt.plot(x_values, y_values, color='darkblue', linestyle='--', linewidth=2, label='Decision Boundary')
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Generated Datasets A and B')
plt.show()
```

Find the optimal decision boundary for the classification of the dataset A and B using Σ1 = Σ2=
\begin{bmatrix} 0.6 & 0.25 \\ 0.25 & 0.4 \end{bmatrix}


Plot the dataset A and B with different colors and plot the obtained optimal
decision boundary. Comment on the characteristics of obtained decision boundary.


---



```python
n = 2000
mu1 = np.array([0, 0])
mu2 = np.array([2, 2])
sigma = np.array([[0.6, 0.25],[0.25, 0.4]])

# Generate dataset A
np.random.seed(6)  # For reproducibility
data_A = np.random.multivariate_normal(mu1, sigma, n)

# Generate dataset B
np.random.seed(3)  # For reproducibility
data_B = np.random.multivariate_normal(mu2, sigma, n)
```

```python
sigmaI = np.linalg.inv(sigma)
w=np.dot(sigmaI,(mu1-mu2)).flatten()
b= 0.5 *(np.dot(np.dot(mu1.T, sigmaI),mu1)-np.dot(np.dot(mu2.T,sigmaI),mu2)).flatten()
```

```python
x_values = np.linspace(-2, 4, 500)
y_values = (-w[0] * x_values + b) / w[1]
```

```python
plt.scatter(data_A[:, 0], data_A[:, 1], c='red', marker='.', label='Dataset A',alpha=0.4)
plt.scatter(data_B[:, 0], data_B[:, 1], c='green', marker='.', label='Dataset B',alpha=0.4)
plt.plot(x_values, y_values, color='black', linestyle='--', linewidth=2, label='Decision Boundary')
plt.legend()
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Generated Datasets A and B')
plt.show()
```
